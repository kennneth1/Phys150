```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
from statistics import mean
import math
from math import log
import numpy as np
import matplotlib as mpl


mpl.rcParams['figure.figsize'] = (20,10)
```

```python
#dataset = pd.read_csv(os.path.expanduser("~/Desktop/constituents_csv.csv"))

path = os.path.expanduser(os.path.expanduser("~/Desktop/1000V"))
files1 = []

for r, d, f in os.walk(path):
    for file in f:
        if '.CSV' in file:
            files1.append(os.path.join(r, file))
v1 = []
thresh1 = []
#list.insert(index, element)

###~170,000 pulses detected per trial # MAKE SURE TO HAVE 3 FOLDERS: 1000V, 1200V, etc... on your desktop
for file in files1:
    i=0
    df = pd.read_csv(file)
    df = df['4000'].loc[15:]
    
    
    filtered = df[df.astype(float) <= -.005]  #Applying a thresh hold of -5 mV
    voltages = df.values.tolist() #All the data
    v1.extend(voltages) 
    thresh1.extend(filtered)

v1 = [(float(x)*-1) for x in v1]
thresh1 = [(float(x)*-1) for x in thresh1]
plt.hist(v1, bins=40)
plt.title("Histogram for signals at 1000V, Beta (Sr)")
plt.xlabel("Output signal (mV)")
plt.ylabel("Counts")

```
