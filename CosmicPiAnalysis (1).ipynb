```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import numpy as np


with open((os.path.expanduser("~/Desktop/PHYS150/Kenneth_Output.txt"))) as fin, open((os.path.expanduser("~/Desktop/PHYS150/Kenneth_Output.csv")), 'w') as fout:
    o=csv.writer(fout)
    for line in fin:
        o.writerow(line.split())             

        
df = pd.read_csv(os.path.expanduser("~/Desktop/PHYS150/Kenneth_Output.csv"))
df.columns = ['Pi_ID', 'Seconds', 'Count']
df['Count'] = df['Count']-1
df['Seconds'] = df['Seconds']-65860
df
```

```python
## Add a frequencies column that tracks change in frequencies over time
```

```python
df['Frequency'] = df['Count']/df['Seconds']
df.head(10)
```

```python
## Appears to be a slightly lag-time in picking up detections for the first 100 seconds or so
## After this point, frequency of detections seem to average out to .080~~ Hz
```

```python
fig, ax = plt.subplots(figsize = (15,6))
plt.plot(df['Frequency'])
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency')
plt.title('Frequency over Time (2 SiPM Detectors)')
```
